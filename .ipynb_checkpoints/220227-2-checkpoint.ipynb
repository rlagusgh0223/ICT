```python
# 이미지 다운로드
import requests
from bs4 import BeautifulSoup
url = "https://search.daum.net/search?w=tot&q=2021%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"
res = requests.get(url)
res.raise_for_status()


soup = BeautifulSoup(res.text, "html.parser")
images = soup.find_all("img", attrs={"class":"thumb_img"})
images
```

```python
for idx, image in enumerate(images):
    #print(idx, image)
    img_url = image['src']
    print(idx, img_url)
    
    # 아래에 주소를 누르면 해당 주소로 이동한다
    if img_url.startswith("//"):    # //로 시작한다면
        img_url = "https:" + img_url    # https를 붙이고 url주소를 붙인다
        
    img_res = requests.get(img_url)
    img_res.raise_for_status()
    
    # 폴더에 이미지를 저장시키지만 현재 폴더에 저장되므로 폴더에 저장된 이미지 지웠다
    with open("movie{}.jpg".format(idx+1), "wb") as f:
        f.write(img_res.content)
        
    if idx >= 4:
        break
```

```python
# 모범답안
for year in range(2016, 2022):
    url = "https://search.daum.net/search?w=tot&q={0}%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR".format(year)
    res = requests.get(url)
    res.raise_for_status()

    soup = BeautifulSoup(res.text, "html.parser")
    images = soup.find_all("img", attrs={"class":"thumb_img"})
    
    for idx, image in enumerate(images):
        img_url = image['src']
        print(idx, img_url)

        if img_url.startswith("//"):
            img_url = "https:" + img_url

        img_res = requests.get(img_url)
        img_res.raise_for_status()

        with open(r"C:\Users\Owner\Desktop\khh\SeoulICT\220227\movie{0}_{1}.jpg".format(idx+1, year), "wb") as f:
            f.write(img_res.content)

        if idx >= 4:
            break
```

# 다음뉴스 데이터 크롤링

```python
import requests
import lxml.html

url = "https://search.daum.net/search?w=news&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&p={}"
```

```python
page = 1
url.format(page)
```

```python
res = requests.get(url.format(page))
```

```python
root = lxml.html.fromstring(res.text)
```

```python
res.text
```

```python
for link in root.cssselect('a.f_nb'):
    print(link.text)
```

```python
for link in root.cssselect('a.f_nb'):
    print(link.attrib['href'])
```

```python
urls = []
for link in root.cssselect('span.cont_info a.f_nb'):
    urls.append(link.attrib['href'])
urls
```

```python
urls = []
for page in range(1, 3):
    res = requests.get(url.format(page))
    root = lxml.html.fromstring(res.text)
    for link in root.cssselect('span.cont_info a.f_nb'):
        urls.append(link.attrib['href'])
urls
```

```python
url = urls[0]
res = requests.get(url)
root = lxml.html.fromstring(res.text)
```

```python
div = root.cssselect(".article_view")[0]
div
```

```python
div.text
```

```python
div.text_content()
```

- 본문기사내용(전체 스크래핑)
- 인공지능.xlsx 저장

```python
# 본문에 대한 데이터 수집
texts = []
for url in urls:
    res = requests.get(url)
    root = lxml.html.fromstring(res.text)
    try:
        div = root.cssselect('.article_view')[0]
    except IndexError:    # .article_view가 없는 경우가 있음
        continue
    texts.append(div.text_content())
```

```python
# 엑셀에 데이터 넣기
import pandas as pd
df = pd.DataFrame({'text' : texts})
df.to_excel(r"C:\Users\Owner\Desktop\khh\SeoulICT\220227\인공지능.xlsx")
```
