# 4. 희소 행렬 변환(SparseTensor)
x_train : scipy의 csr 포맷  
model : 텐서플로 포맷(coo)

```python
x_train
```

```python
type(x_train)
```

```python
tf.SparseTensor
```

```python
x_coo = x_train.tocoo()
```

```python
# 행번호
x_coo.row
```

```python
# 열번호
x_coo.col
```

```python
# 내용
x_coo.data
```

```python
# 행번호와 열번호를 컬럼으로 합치기
import numpy as np

index = np.column_stack([x_coo.row, x_coo.col])
index
```

```python
# 텐서플로의 SparseTensor로 변환
x_train_sparse = tf.SparseTensor(index, x_coo.data, x_coo.shape)
```

```python
x_train_sparse = tf.sparse.reorder(x_train_sparse)
```

```python
model.fit(x_train_sparse, y_train, epochs=2)
```
