## 1.2 데이터 준비

```python
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))    # 세로는 len만큼, 가로는 dimension만큼
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.    # 소수점으로 맞추기 위해 1.으로 입력
    print()
    return results

x_train = vectorize_sequences(train_data).astype('float32')
x_test = vectorize_sequences(test_data).astype('float32')
```

```python
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

## 1.3. 신경망 모델 만들기

```python
from tensorflow.keras import models, layers

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
# 방법1
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 방법2
# learning_rate 지정
from tensorflow.keras import optimizers

model.compile(optimizer=optimizers.RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras import losses, metrics, optimizers

model.compile(optimizer=optimizers.RMSprop(learning_rate=0.001), loss=losses.binary_crossentropy, metrics=[metrics.binary_accuracy])
```

## 1.4 훈련 검증

```python
x_val = x_train[:10000].astype(float)    # 혹시 int인 데이터가 있을 수 있으니까 float으로 만들어준다
partial_x_train = x_train[10000:].astype(float)

y_val = y_train[:10000].astype(float)
partial_y_train = y_train[10000:].astype(float)
```

```python
history = model.fit(partial_x_train, partial_y_train, epochs=20, batch_size=512, validation_data=(x_val, y_val))
```

```python
loss = history.history['loss']    # 매순간의 loss값 넘김
val_loss = history.history['val_loss']
acc = history.history['binary_accuracy']
val_acc = history.history['val_binary_accuracy']
```

```python
import matplotlib.pyplot as plt

acc = history.history['binary_accuracy']
val_acc = history.history['val_binary_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)
print(epochs)
print(loss)

# 'bo'는 파란색 점을 의미
plt.plot(epochs, loss, 'bo', label='Training loss')    # 오차율
# 'b'는 파란색 실선 의미
plt.plot(epochs, val_loss, 'b', label='Validation loss')    # 검증 데이터
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()    # 그래프를 보면 4 정도에서 끊어야 오버피팅이 일어나지 않을거라고 예상할 수 있다
```

```python
# accuracy 시각화

plt.plot(epochs, acc, 'bo', label='Training accuracy')    # 오차율
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()    # 그래프를 보면 4 정도에서 끊어야 오버피팅이 일어나지 않을거라고 예상할 수 있다
```

```python
# 위의 검증을 통해 epochs를 4에서 끊어봄
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=4, batch_size=512)
results = model.evaluate(x_test, y_test)
```
